```python
from sklearn.datasets import make_blobs
from gmm import GMM
from k_means import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

## Create dataset

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=10, cluster_std=1.3,
                  random_state=20)
```

```python
plt.scatter(X[:, 0], X[:, 1], c='grey', s=30)
plt.axis('equal')
plt.show()
```

## K-means clustering

```python
k = KMeans(n_centroids=20, n_iters=100)
k.fit(X)
```

```python
clustered = k.assigned
for ii in clustered:
      plt.scatter(ii[:,0], ii[:,1])
plt.show()        
```

## Gaussian Mixture Model 

```python
gmm = GMM(C=10, rtol=1e-3, max_iter=200, restarts=5)
gmm.fit(X)
```

```python
labels = gmm.gamma.argmax(axis=1)
for c in range(20):
    x_tmp = X[labels==c]
    plt.scatter(x_tmp[:, 0], x_tmp[:, 1])
plt.show()    
```
